```python
# Data reading function
def read_valid_json_files(folder_path):
    json_files = [file for file in os.listdir(folder_path) if file.endswith('.json')]
    valid_data = []
    for json_file in json_files:
        try:
            with open(os.path.join(folder_path, json_file), 'r') as f:
                data = json.load(f)
                valid_data.extend(data)
        except json.JSONDecodeError:
            print(f"Skipping invalid JSON file: {json_file}")
    return valid_data
```

```python
# Data extracting function
def extract_data(data):
    extracted_data = []
    i = 0
    for item in data:
        user_id = item.get("user_id")
        text = item.get("text")
        username = item.get("username")
        file_number = i+1
        if file_number >= 6:
          i = 0
        else:
          i += 1
        extracted_data.append({"username": username, "user_id": user_id, "text": text, "file_number":file_number})
    return extracted_data
```

```python
import nltk
import re
import emoji

nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")

from nltk.corpus import stopwords
import string

stopword=set(stopwords.words('english'))

def clean(text):
    # Chuy·ªÉn ƒë·ªïi vƒÉn b·∫£n th√†nh ch·ªØ th∆∞·ªùng
    text = str(text).lower()
    # Lo·∫°i b·ªè c√°c k√Ω t·ª± trong d·∫•u ngo·∫∑c vu√¥ng
    text = re.sub('\[.*?\]', '', text)
    # Lo·∫°i b·ªè c√°c ƒë∆∞·ªùng d·∫´n web
    text = re.sub('https?://\S+|www\.\S+', '', text)
    # Lo·∫°i b·ªè c√°c th·∫ª HTML
    text = re.sub('<.*?>+', '', text)
    # Lo·∫°i b·ªè c√°c d·∫•u c√¢u
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    # Lo·∫°i b·ªè c√°c k√Ω t·ª± xu·ªëng d√≤ng
    text = re.sub('\n', '', text)
    # Lo·∫°i b·ªè c√°c t·ª´ c√≥ ch·ª©a s·ªë
    text = re.sub('\w*\d\w*', '', text)
    # Lo·∫°i b·ªè emoji
    text = emoji.replace_emoji(text)
    # Lo·∫°i b·ªè c√°c t·ª´ d·ª´ng, v√≠ d·ª• nh∆∞ l√†, c·ªßa, v√†.
    # T·ª´ d·ª´ng l√† nh·ªØng t·ª´ ph·ªï bi·∫øn nh∆∞ng kh√¥ng mang nhi·ªÅu √Ω nghƒ©a cho vƒÉn b·∫£n.
    text = [word for word in text.split(' ') if word not in stopword]
    # N·ªëi l·∫°i c√°c t·ª´ c√≤n l·∫°i th√†nh m·ªôt chu·ªói vƒÉn b·∫£n
    text = " ".join(text)
    # √Åp d·ª•ng thu·∫≠t to√°n stemmer ƒë·ªÉ r√∫t g·ªçn c√°c t·ª´ v·ªÅ d·∫°ng g·ªëc,
    # v√≠ d·ª• nh∆∞ h·ªçc -> h·ªçc, h·ªçc sinh -> h·ªçc, h·ªçc t·∫≠p -> h·ªçc.
    # Stemmer l√† m·ªôt c√¥ng c·ª• ƒë·ªÉ gi·∫£m thi·ªÉu s·ªë l∆∞·ª£ng c√°c t·ª´ kh√°c nhau nh∆∞ng c√≥ c√πng nghƒ©a trong vƒÉn b·∫£n.
    text = [stemmer.stem(word) for word in text.split(' ')]
    # N·ªëi l·∫°i c√°c t·ª´ ƒë√£ ƒë∆∞·ª£c stem th√†nh m·ªôt chu·ªói vƒÉn b·∫£n
    text = " ".join(text)
    return text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(data_fb['text'])

# Compute similarity scores
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
nltk.download('punkt')

# Tokenize and train Word2Vec model
tokenized_text = data_fb['text'].apply(word_tokenize)
word2vec_model = Word2Vec(sentences=tokenized_text, vector_size=100, window=5, min_count=1, workers=4)
```

```python
def embed_text(text):
    # Create an embedding for a given text by averaging word embeddings
    words = word_tokenize(text)
    word_embeddings = [word2vec_model.wv[word] for word in words if word in word2vec_model.wv]

    if not word_embeddings:
        return np.zeros(word2vec_model.vector_size)

    return np.mean(word_embeddings, axis=0)
```

```python
# Add word embeddings as a new feature
data_fb['text_embeddings'] = data_fb['text'].apply(embed_text)
```

```python
import math
def cosine(v1,v2):
    "compute cosine similarity of v1 to v2: (v1 dot v2)/{||v1||*||v2||)"
    sumxx, sumxy, sumyy = 0, 0, 0
    for i in range(len(v1)):
        x = v1[i]; y = v2[i]
        sumxx += x*x
        sumyy += y*y
        sumxy += x*y
    return sumxy/math.sqrt(sumxx*sumyy)
```

# **6. MODEL**

```python
import math
import re
from collections import Counter
import pandas as pd

WORD = re.compile(r"\w+")


def get_cosine(vec1, vec2):
    intersection = set(vec1.keys()) & set(vec2.keys())
    numerator = sum([vec1[x] * vec2[x] for x in intersection])

    sum1 = sum([vec1[x] ** 2 for x in list(vec1.keys())])
    sum2 = sum([vec2[x] ** 2 for x in list(vec2.keys())])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)

    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator


def text_to_vector(text):
    words = WORD.findall(text)
    return Counter(words)
```

```python
import pke

data_kw_fb = pd.DataFrame(columns=['fb_text', 'fb_keyphrase','fb_score'])
for i in range(0,30):
    extractor = pke.unsupervised.TopicRank()
    extractor.load_document(input=str(data_fb['text'][i]),
                            language='en',
                            normalization='stemming')
    extractor.candidate_selection(pos={"NOUN", "PROPN","ADJ"})
    ex = extractor.candidate_weighting(threshold=0.74,
                                method="average")
    fb_keyphrase=[]
    fb_score=[]
    for (keyphrase, score) in extractor.get_n_best(n=20, stemming=True):
        fb_keyphrase.append(keyphrase)
        fb_score.append(score)
    data_kw_fb = data_kw_fb.append({
                    'fb_text': data_fb['text'][i],
                    'fb_keyphrase':fb_keyphrase ,
                    'fb_score':fb_score
                }, ignore_index=True)
```
